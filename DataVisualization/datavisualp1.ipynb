#### Stack Plot
    the idea of stack plots is to show "parts to the whole" over time. A stack plot is basically like a pie-chart, only over time.
    
Let's consider a situation where we have 24 hours in a day, and we'd like to see how we're spending our time. We'll divide our activities into: Sleeping, eating, working, and playing.

### no labels in stakplot so we do a work

```python
# data 
days = [1,2,3,4,5]

#categorizing data
sleeping = [7,8,6,11,7]
eating =   [2,3,4,3,2]
working =  [7,8,7,2,2]
playing =  [8,5,7,8,13]
#our "x" axis will consist of the days variable, which is 1,2,3,4, and 5. 

#making our own individual labels (draws a new plot on top of stakplot )
plt.plot([],[],color='m', label='Sleeping', linewidth=5)
plt.plot([],[],color='c', label='Eating', linewidth=5)
plt.plot([],[],color='r', label='Working', linewidth=5)
plt.plot([],[],color='k', label='Playing', linewidth=5)

#use stakplot to show, pass colors for different items
plt.stackplot(days, sleeping,eating,working,playing, colors=['m','c','r','k'])


#labels and show
plt.xlabel('x')
plt.ylabel('y')
plt.title('Interesting Graph\nCheck it out')
plt.show()
```

### Pie Charts
    Pie charts are a lot like the stack plots, only they are for a certain point in time. Typically, a Pie Chart is used to show parts to the whole, and often a % share. Luckily for us, Matplotlib handles the sizes of the slices and everything, we just feed it the numbers.

to make pie chart use plt.pie()
shadow is used to give it shadow
explode is used to bring a slice out a bit
autopct adds percentage to pie

```python
slices = [7,2,2,13]
activities = ['sleeping','eating','working','playing']
cols = ['c','m','r','b']

plt.pie(slices,
        labels=activities,
        colors=cols,
        startangle=90,
        shadow= True,
        explode=(0,0.1,0,0),
        autopct='%1.1f%%')

plt.title('Interesting Graph\nCheck it out')
plt.show()
```

#### Loading data from files into displaying a graph
using example.txt file 
containing 
some numbers data 

```python
# import csv to read file
import csv

#create list to load data in
x = []
y = []

# open file read text and add to x and y 
with open('example.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    for row in plots:
        x.append(int(row[0]))
        y.append(int(row[1]))

# plot the points
plt.plot(x,y, label='Loaded from file!')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Interesting Graph\nCheck it out')
plt.legend()
plt.show()
```

#### another way we can read the file is using numpy lets do an example of that


```python
#import numpy
import numpy as np

# load txt into x,y using numpy
# numpy makes it easiet to load it in
# unpacking variables using delimeter
x, y = np.loadtxt('example.txt', delimiter=',', unpack=True)

# plot 
plt.plot(x,y, label='Loaded from file using numpy!')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Interesting Graph\nCheck it out')
plt.legend()
plt.show()
```

#### Coverting the data 
        Converting date stamps for Matplotlib
 To do this, we're going to write a new function, bytespdate2num.

```python
# getting the data from the internet 

# urlib to load url , and importing dates
import urllib
# handles mdates
import matplotlib.dates as mdates

#lets define a function to get the stocks data

#fmt = format of date
#encoding type
# not neccassary to understand this function
def bytespdate2num(fmt, encoding='utf-8'):
    strconverter = mdates.strpdate2num(fmt)
    def bytesconverter(b):
        s = b.decode(encoding)
        return strconverter(s)
    return bytesconverter

def graph_data(stock):

    # url of data
    stock_price_url = 'https://pythonprogramming.net/yahoo_finance_replacement'
    # reading and decoding data
    source_code = urllib.request.urlopen(stock_price_url).read().decode()
    # empty list for data
    stock_data = []
    # spliting data into rows
    # split by new lines
    split_source = source_code.split('\n')

    for line in split_source[1:]:
        #split by coma
        split_line = line.split(',')
        # if vals in line 7 
        if len(split_line) == 7:
            if 'values' not in line:
                #appends it to list
                stock_data.append(line)
    #What we are doing here, is unpacking these six elements to six variables, 
    #with numpy's loadtxt function. 
    #he first parameter here is stock_data, 
    #which is the data we're loading. Then, we specify the delimiter, 
    #which is a comma in this case, then we specify that we indeed want to unpack the variables here not just to one variable,
    #but to this group of variables we've defined. Finally, 
    #we use the optional "converters" parameter to specify what element we want to convert (0), 
    #and then how we want to do that. We pass a function called bytespdate2num, 
    #which doesn't quite exist yet, but we'll write that next.
    date, closep, highp, lowp, openp, adj_closep, volume = np.loadtxt(stock_data,
                                                          delimiter=',',
                                                          unpack=True,
                                                          # %Y = full year. 2015
                                                          # %y = partial year 15
                                                          # %m = number month
                                                          # %d = number day
                                                          # %H = hours
                                                          # %M = minutes
                                                          # %S = seconds
                                                          # 12-06-2014
                                                          # %m-%d-%Y
                                                          converters={0: bytespdate2num('%Y-%m-%d')})
    
    #plotting
    plt.plot_date(date, closep,'-', label='Price')

    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.title('Interesting Graph\nCheck it out')
    plt.legend()
    plt.show()
    
#run the function    
graph_data('TSLA')
```
